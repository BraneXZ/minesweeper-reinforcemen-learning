```
# Test to make sure the minesweeper class works correctly
def test_minesweeper(row, col, mine, n=1000):
    wins = 0
    
    for i in range(n):
        ms = MineSweeper(row, col, mine)
        visited = set()
        ms.action(val=random.randint(0, row * col - 1))

        while ms.unknown_fields > 0:
            rand_field = random.sample(set(range(ms.row * ms.col)) - ms.mine_location - visited, 1)[0]
            visited.add(rand_field)
            ms.action(val=rand_field)

        if ms.win:
            wins += 1
    print(wins)
```
