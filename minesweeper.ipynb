```
import gym
```

```
class MineSweeperEnv(gym.Env):
    def __init__(self, row, col, mine):
        self.minesweeper = MineSweeper(row, col, mine)
        self.action_space = gym.spaces.Discrete(row*col)
        # self.observation_space = gym.spaces.Discrete(2)

    def step(self, action):
        reward = self.minesweeper.action(val=action)
            
        # regardless of the action, game is done after a single step
        if self.minesweeper.win or reward == -1:
            done = True
        else:
            done = False

        info = {}

        return self.minesweeper.board, reward, done, info

    def reset(self):
        return self.minesweeper.reset_board()
```

```
env = MineSweeperEnv(3, 3, 3)
```
